# Underfitting and Overfitting

**Overfitting** <br> <br>
Kondisi dalam machine learning ketika sebuah model terlalu “menghafal” data latih, termasuk noise, pola acak, dan hal-hal yang sebenarnya tidak penting.<br> Akibatnya, model tampil sangat baik pada data training tetapi buruk saat digunakan pada data baru, karena ia tidak benar-benar memahami pola yang berlaku secara umum. <br> <br> 

Sebuah model cocok dengan data pelatihan hampir secara sempurna, tetapi berkinerja buruk pada data validasi maupun data baru lainnya. Sebaliknya, jika kita membuat pohon keputusan (decision tree) sangat dangkal, model tidak dapat membagi class ke dalam kelompok-kelompok yang benar-benar berbeda. <br> <br>
**Underfitting** <br>

Pada kondisi ekstrem, jika sebuah tree hanya membagi data menjadi 2 atau 4 kelompok, setiap kelompok masih berisi berbagai macam item yang berbeda. Akibatnya, prediksi bisa meleset jauh untuk sebagian besar item, bahkan pada data pelatihan (dan akan buruk juga pada data validasi karena alasan yang sama). Ketika sebuah model gagal menangkap perbedaan dan pola penting dalam data sehingga kinerjanya buruk bahkan pada data pelatihan, kondisi itu disebut underfitting.

<br> <br>
Kita ingin mencari titik tengah dari Overfitting dan Underfitting agar membuat model yang baik

````python
# Example Code 

from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)

# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))

````

### Conclusion

**Overfitting**: menangkap pola-pola acak yang tidak akan muncul lagi di masa depan, sehingga membuat prediksi menjadi kurang akurat. <br><br>

**Underfitting**: gagal menangkap pola-pola yang relevan, yang juga menyebabkan prediksi menjadi kurang akurat.

# Random Forest

Random forest menggunakan banyak pohon keputusan (decision tree).<br>
Setiap pohon membuat prediksi sendiri, lalu random forest mengambil rata-rata dari semua prediksi tersebut.<br>
Karena menggabungkan banyak pohon, hasilnya biasanya jauh lebih akurat dibanding hanya memakai 1 decision tree saja.<br>

````python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
````
